```
fight_map = {} #takes game id as input, return whether terran vs terran, terran vs zerg, etc
game_map = {'tt':[], 'tz':[], 'pt':[], 'pz':[], 'pp':[], 'zz':[]} # takes tvst, tvsz as input, return game id
species = {'T':1, 'Z':2, 'P':3}
```

```
def map_fight(df, fight_map):
  for index, row in df.iterrows():
    game_id = row.game_id
    if game_id in fight_map and np.all(fight_map[game_id]): continue # both player record recorded
    if game_id not in fight_map: # both player record not recorded
      fight_map[game_id] = [0,0]
      fight_map[game_id][row.player] = species[row.species]
    else: # only only player record not recorded
      fight_map[game_id][row.player] = species[row.species]
  return fight_map
```

```
map_fight(train, fight_map)
```

```
  result = pd.DataFrame.from_dict(fight_map)
```

```
result.to_csv('drive/My Drive/result.csv')
# game_map = {'tt':[], 'tz':[], 'pt':[], 'pz':[], 'pp':[], 'zz':[]} # takes tvst, tvsz as input, return game id
# species = {'T':1, 'Z':2, 'P':3}
```

```
for game_id in fight_map:
  if fight_map[game_id] == [1,1]:
    game_map['tt'].append(game_id)
  elif fight_map[game_id] == [1,2] or fight_map[game_id] == [2,1]:
    game_map['tz'].append(game_id)
  elif fight_map[game_id] == [1,3] or fight_map[game_id] == [3,1]:
    game_map['pt'].append(game_id)
  elif fight_map[game_id] == [2,3] or fight_map[game_id] == [3,2,]:
    game_map['pz'].append(game_id)
  elif fight_map[game_id] == [2,2]:
    game_map['zz'].append(game_id)
  elif fight_map[game_id] == [3,3]:
    game_map['pp'].append(game_id)

    
```

```
game_result = pd.DataFrame.from_dict(game_map, orient='index')
game_result.to_csv('drive/My Drive/dacon/game_result.csv')
```
